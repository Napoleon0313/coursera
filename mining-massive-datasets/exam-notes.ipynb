# Chapter 2 - Map Reduce

Much of the course will be devoted to large scale computing for data mining

Challenges: 
- How to distribute computation
- Distributed / paralell programming hard

Map reduce addresses these challanges:
- Googles computational/data manipulation model
- Elegang way to work with big data

Large scale computing for data-mining challanges on commodity hardware:
- How do we distribute the computation?
- How can we make it easy to write a distributed program?
- Machines will fail

Problem: copying data over a network takes time
- Bring computation close to the data
- Store files multiple times for reliability

Map reduce:
- Storage infrastructure - File system
    - Google: GFS, Hadoop: HDFS
- Programming model
    - Map reduce
    
Overview:
- Sequentally load a lot of data
- ** MAP **
    - Extract something you care about
- ** SORT **
    - Sort and shuffle
- ** REDUCE **
    - Aggregate, summarize, filter or transform
- ... write the result.

The map reduce environment takes care of:

- Partitioning
- Scheduling
- Group by key 
- Handling machine failures
- Managing required inter-machine communication

The master nodes takes care of: 
- Task status
- Idle tasks
- When a map task completes, it sends the master location and sizes to of its R intermediary files
- Master pings workers periodically to detect failures

How many map and reduce jobs?
- M much larger than the nodes in the cluster
- Usually R is smaller than M

Map-reduce join:
- Each	input	tuple	R(a,b)	into	key-value	pair	(b,(a,R))	
- Each	input	tuple	S(b,c)	into	(b,(c,S))	
- Each	Reduce	process	matches	all	the	pairs	(b,(a,R))	with	all	(b,(c,S))	and	outputs	(a,b,c).	

** In map reduce we quantify the cost of running an algorithm **
- Communication cost = total I/O of the whole process
- Elapsed communication cost = max I/O along any path
- Computational cost = count running time of processes

** Example of cost measures **
- Communica-on	cost	= input	file	size	+	2	×	(sum	of	the	sizes	of	all	files	passed	from Map	processes	to	Reduce	processes)	+	the	sum	of	the	output	sizes	of	the	Reduce	processes.	
-  Elapsed	communica-on	cost is	the	sum	of	the	largest	input	+	output	for	any	map	process,	plus	 the	same	for	any	reduce	process	

** What does cost measure mean? **
- Either	the	I/O	(communica6on)	or	processing	(computation)	cost	dominates. Ignore	one	or	the	other	
-  Total	cost	tells	what	you	pay	in	rent	from		your	friendly	neighborhood	cloud	
-  Elapsed	cost	is	wall-clock	time	using	parallelism	

# Finding Similar Items

> Many problems can be expressed as finding “similar” sets

Examples:
- Finding pages with similar words
- Customers who purchased similar products
- Images with similar features

Given high dimensional data points x1 and x2
- For example an image

And a distance function d(x1, x2)

Find all pairs of the data that are within some distance treshold d(x1, x2) <= s

How can this be done in O(n)?

Jaccard similarity = A intersection B / A union B

** Essential steps for similar docs **
1. Shingling - convert documents to sets
2. Min-hashing - convert large sets to short signatures, while preserving similarity
3. Locality-sensitivity-hashing: Focus on signatures likely to be from similar documents
